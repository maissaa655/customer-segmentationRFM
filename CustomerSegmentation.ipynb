<h3><font color=#d4537e>📌 Business Context</font></h3>

Understanding **who your customers are** is the foundation of any effective marketing strategy. In a competitive retail landscape, treating all customers the same leads to wasted budgets, poor retention, and missed revenue opportunities. The goal of this project is to move from a one-size-fits-all approach to a **data-driven segmentation** that reveals distinct customer profiles — each with its own behaviour, value, and engagement pattern.

<h3><font color=#d4537e>📦 Dataset</font></h3>

The analysis is built on the **UCI Online Retail Dataset**, a real-world transactional record from a UK-based non-store online retailer specialising in gift and home décor products. The dataset spans **December 2010 to December 2011** and contains over **540,000 transactions** across **38 countries**, with the United Kingdom accounting for approximately **91%** of all activity.

| Field | Description |
|:---|:---|
| `InvoiceNo` | Unique transaction identifier (prefix `C` = cancellation) |
| `StockCode` | Unique product identifier |
| `Description` | Product name |
| `Quantity` | Units purchased per transaction line |
| `InvoiceDate` | Date and time of transaction |
| `UnitPrice` | Price per unit (GBP £) |
| `CustomerID` | Unique customer identifier |
| `Country` | Customer country of residence |

<h3><font color=#d4537e>🎯 Objectives</font></h3>

1. **Engineer meaningful customer-level features** — going beyond simple RFM (Recency, Frequency, Monetary) to include tenure, behavioural patterns, product diversity, and geographic signals.
2. **Clean and prepare the data rigorously** — handling cancellations, missing identifiers, anomalous stock codes, and zero-price transactions.
3. **Apply unsupervised clustering** — using K-Means on PCA-reduced features to identify stable, well-separated customer segments.
4. **Interpret and profile each segment** — translating mathematical clusters into actionable business personas with clear strategic recommendations.

# <b><span style='color:#ff69b4'> 2. Exploratory Data Analysis (EDA) </span></b>

## <b><span style='color:#ff69b4'> 2.1 Univariate Analysis </span></b>

### <b><span style='color:#ff69b4'> 2.1.1 Distribution of Numerical Variables (Quantity, UnitPrice) </span></b>

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

for i, col in enumerate(['Quantity', 'UnitPrice']):
    data = df[col].dropna()

    # Couper plus agressivement : 5% - 95%
    q_low  = data.quantile(0.05)
    q_high = data.quantile(0.95)
    data_clipped = data[(data >= q_low) & (data <= q_high)]

    # --- Histogram — Raw Distribution ---
    ax = axes[i, 0]
    ax.hist(data_clipped, bins=50, color='#7f77dc', alpha=0.85, edgecolor='white', linewidth=0.3)
    ax.set_title(f'{col} — Raw Distribution (5th–95th percentile)')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

    # --- Boxplot — données clippées pour mieux voir la boîte ---
    ax = axes[i, 1]
    ax.boxplot(data_clipped, vert=False, patch_artist=True,
               boxprops=dict(facecolor='#64557e', alpha=0.6),
               flierprops=dict(marker='.', markersize=2, alpha=0.3))
    ax.set_title(f'{col} — Boxplot (5th–95th percentile)')
    ax.set_xlabel(col)

    # --- Log-transformed (toutes les valeurs positives) ---
    ax = axes[i, 2]
    pos = data[data > 0]
    ax.hist(np.log1p(pos), bins=60, color='#f84e75', alpha=0.85, edgecolor='white', linewidth=0.3)
    ax.set_title(f'log1p({col}) — Normalized View')
    ax.set_xlabel(f'log1p({col})')
    ax.set_ylabel('Frequency')

    # Stats
    print(f'\n{col}:')
    print(f'  Mean            : {data.mean():.2f}')
    print(f'  Median          : {data.median():.2f}')
    print(f'  Std Dev         : {data.std():.2f}')
    print(f'  Skewness        : {data.skew():.2f}')
    print(f'  Kurtosis        : {data.kurtosis():.2f}')
    print(f'  Negative values : {(data < 0).sum()}')
    print(f'  Zero values     : {(data == 0).sum()}')
    print(f'  Min / Max       : {data.min():.2f} / {data.max():.2f}')
    print(f'  Clipped range   : {q_low:.2f} → {q_high:.2f}')

plt.tight_layout()
plt.show()
```

<div style="border-radius:10px; padding: 15px; background-color: #ffeaef; font-size:120%; text-align:left">


- __`Quantity`__:
   - The distribution is extremely right-skewed, with the vast majority of transactions clustered near zero, confirming that most purchases involve small quantities.
   - The boxplot reveals massive outliers on both sides — negative values (returns/cancellations) and extreme positive values (bulk orders) — making the raw distribution unreadable.
   - After `log1p` transformation, the distribution normalizes into a clear bell shape centered around 1–2, showing that typical transactions involve 3 to 7 units.
   - **10,624 negative values** (returns) and **2 zero values** must be filtered before building customer features.

- __`UnitPrice`__:
   - Same pattern as Quantity — the raw distribution is dominated by a spike near zero, with extreme outliers stretching the x-axis to 40,000.
   - The boxplot confirms that nearly all prices cluster between 0 and 5 £, with a handful of extreme values distorting the mean (4.61 £) far above the median.
   - After `log1p` transformation, the distribution reveals a realistic price range concentrated between 1 and 4 (i.e., roughly 1.7 £ to 54 £), with a long but manageable right tail.
   - **2 negative values** and **zero prices** are present and must be removed — they do not represent real transactions.

```python
# Temporal distributions
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Hour']        = df['InvoiceDate'].dt.hour
df['DayOfWeek']   = df['InvoiceDate'].dt.dayofweek
df['Month']       = df['InvoiceDate'].dt.month

day_labels   = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
month_labels = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

hour_counts = df['Hour'].value_counts().sort_index()
axes[0].bar(hour_counts.index, hour_counts.values, color='#7f77dd', alpha=0.85)
axes[0].set_title('Transactions by Hour of Day')
axes[0].set_xlabel('Hour')
axes[0].set_ylabel('Count')

dow_counts = df['DayOfWeek'].value_counts().sort_index()
axes[1].bar([day_labels[i] for i in dow_counts.index], dow_counts.values, color='#d4537e', alpha=0.85)
axes[1].set_title('Transactions by Day of Week')
axes[1].set_xlabel('Day')

month_counts = df['Month'].value_counts().sort_index()
axes[2].bar([month_labels[i-1] for i in month_counts.index], month_counts.values, color='#1d9e75', alpha=0.85)
axes[2].set_title('Transactions by Month')
axes[2].set_xlabel('Month')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

- __`Hour of Day`__:

   - For segmentation, **shopping hour will serve as a behavioral signal**: customers who consistently order at unusual hours (early morning or late evening) likely represent a distinct high-value B2B segment placing large bulk orders outside staff hours.

- __`Day of Week`__:
   - Activity is concentrated **Monday through Thursday**, with Sunday being the weakest day by far.

      - **Year-round buyers** — consistently active across all months, representing the retailer's loyal core customer base.
   - Recency scores will be heavily influenced by this seasonality: a customer last seen in November may have a high recency value not because they churned, but because they are a **seasonal buyer** — a nuance the segmentation model must account for.

## <b><span style='color:#ff69b4'> 2.2 Bivariate Analysis </span></b>

### <b><span style='color:#ff69b4'> 2.2.1 Correlation Matrix </span></b>

```python

# ── Pink colormap ────────────────────────────────────────────────────
pink_cmap = mcolors.LinearSegmentedColormap.from_list(
    'pink_map', ['#ffffff', '#f4c0d1', '#d4537e', '#4b1528']
)

# ── Only original dataset numeric features ───────────────────────────
corr = df[['Quantity', 'UnitPrice']].corr(method='spearman')

# ── Plot ─────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(5, 4))

sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    cmap=pink_cmap,
    center=0,
    vmin=-1,
    vmax=1,
    linewidths=0.5,
    linecolor='white',
    annot_kws={'size': 12, 'weight': '500'},
    cbar_kws={'shrink': 0.8, 'label': 'Spearman correlation'},
    ax=ax
)

ax.set_title('Correlation — Quantity vs UnitPrice',
             fontsize=13, fontweight='500', pad=15)
ax.tick_params(axis='x', rotation=0, labelsize=11)
ax.tick_params(axis='y', rotation=0, labelsize=11)

plt.tight_layout()
plt.show()
```

<div style="border-radius:10px; padding: 15px; background-color: #ffeaef; font-size:120%; text-align:left">
<h3 align="left"><font color=#d4537e>Inferences:</font></h3>

- __`Spearman Correlation = -0.39`__:
   - There is a **moderate negative correlation** between `Quantity` and `UnitPrice` — as the quantity purchased increases, the unit price tends to decrease, and vice versa.
   - This is consistent with typical retail behavior: **high-volume items are usually cheaper** (bulk, everyday goods), while **expensive items are purchased in smaller quantities** (premium or specialty products).

- __`What this means for customer profiles`__:
   - Customers who buy in **large volumes** tend to spend less per unit — they are bulk buyers driven by quantity, not product value.
   - Customers who spend **more per transaction** tend to buy fewer units — they are selective, premium-oriented buyers who prioritize quality over volume.
   - These are **two fundamentally different purchasing behaviors** that cannot be captured by a single metric — both the volume dimension and the value dimension must be preserved separately to build meaningful customer profiles..

- __`Segmentation Implication`__:
   - This correlation suggests that **volume-driven clusters** and **value-driven clusters** will naturally emerge in the segmentation — bulk buyers vs. premium buyers — making `Quantity` and `UnitPrice` complementary signals for distinguishing customer profiles.
</div>

### <b><span style='color:#ff69b4'> 2.2.2 Relationships Between Variables </span></b>

```python

clean = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)].copy()
clean['TotalValue'] = clean['Quantity'] * clean['UnitPrice']

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# log-log scatter: Quantity vs UnitPrice
axes[0].scatter(np.log1p(clean['Quantity']), np.log1p(clean['UnitPrice']),
                alpha=0.04, color='#d4537e', s=3)
axes[0].set_title('log1p(Quantity) vs log1p(UnitPrice)')
axes[0].set_xlabel('log1p(Quantity)')
axes[0].set_ylabel('log1p(UnitPrice)')

# Revenue by day of week
clean['DayOfWeek'] = clean['InvoiceDate'].dt.dayofweek
dow_rev = clean.groupby('DayOfWeek')['TotalValue'].sum()
day_labels = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
day_map = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
axes[1].bar([day_map[i] for i in dow_rev.index], dow_rev.values / 1000, color='#1d9e75', alpha=0.85)
axes[1].set_title('Total Revenue by Day of Week (£k)')
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Revenue (£k)')

plt.tight_layout()
plt.show()

# Monthly revenue trend
clean['YearMonth'] = clean['InvoiceDate'].dt.to_period('M')
monthly_rev = clean.groupby('YearMonth')['TotalValue'].sum().reset_index()
monthly_rev['YM_str'] = monthly_rev['YearMonth'].astype(str)

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(monthly_rev['YM_str'], monthly_rev['TotalValue'] / 1000,
        marker='o', color='#d4537e', linewidth=2.5)
ax.fill_between(range(len(monthly_rev)), monthly_rev['TotalValue'] / 1000,
                alpha=0.1, color='#d4537e')
ax.set_title('Monthly Revenue Trend (£k)', fontsize=13)
ax.set_xlabel('Month')
ax.set_ylabel('Revenue (£k)')
ax.tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

# <b><span style='color:#ff69b4'> 4. Feature Engineering </span></b>

## <b><span style='color:#ff69b4'> 4.1 RFM Feature Construction </span></b>

### <b><span style='color:#ff69b4'> 4.1.1 Recency </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
In this step, we aim to measure how recently each customer has interacted with the business through a purchase. Recency is a key indicator of customer engagement and plays a vital role in customer segmentation. For this purpose, I will be constructing the following feature:

RECENCY: This feature captures the number of days elapsed since a customer's most recent transaction. Customers with a smaller number of days since their last purchase are considered more active and engaged, while those with a larger number of days may be drifting away or becoming inactive. Leveraging this information allows businesses to design targeted campaigns aimed at winning back lapsed customers, ultimately improving retention rates and building long-term customer loyalty.

```python
# Convert InvoiceDate to datetime and extract only the date
df['PurchaseDay'] = df['InvoiceDate'].dt.date

# Find the most recent date in the entire dataset
latest_date = pd.to_datetime(df['PurchaseDay'].max())

# Build the customer summary
purchase_summary = df.groupby('CustomerID').agg(
    Country=('Country', 'first'),
    LastPurchaseDay=('PurchaseDay', 'max')
).reset_index()

# Convert LastPurchaseDay to datetime type before subtraction
purchase_summary['LastPurchaseDay'] = pd.to_datetime(purchase_summary['LastPurchaseDay'])

# Calculate the number of days since the last purchase for each customer
purchase_summary['Recency'] = (latest_date - purchase_summary['LastPurchaseDay']).dt.days

# Remove the LastPurchaseDay column
purchase_summary.drop(columns=['LastPurchaseDay'], inplace=True)

print(purchase_summary)
```

### <b><span style='color:#ff69b4'> 4.1.2 Frequency </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
In this step, we will engineer two features that capture how frequently a customer engages in purchasing activity with the retailer:

Total Transactions: serves as a direct measure of how actively a customer interacts with the retailer, with higher values reflecting a stronger and more consistent purchasing habit.
Total Products Purchased: provides a deeper look into the customer's consumption pattern, revealing not just how often they buy, but how much they buy in terms of volume.

Together, these features form the foundation of frequency-based customer segmentation, enabling businesses to distinguish between occasional shoppers and loyal high-volume buyers, which is essential for crafting effective marketing campaigns and delivering personalized shopping experiences

```python
invoice_count = df.groupby('CustomerID')['InvoiceNo'].nunique().reset_index()
invoice_count.rename(columns={'InvoiceNo': 'Total_orders'}, inplace=True)

quantity_sum = df.groupby('CustomerID')['Quantity'].sum().reset_index()
quantity_sum.rename(columns={'Quantity': 'Total_Purchase'}, inplace=True)

purchase_summary = purchase_summary.merge(invoice_count, on='CustomerID', how='left')
purchase_summary = purchase_summary.merge(quantity_sum, on='CustomerID', how='left')

purchase_summary.head()
```

### <b><span style='color:#ff69b4'> 4.1.3 Monetary </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
In this step, we will construct two features that reflect the monetary value of each customer's purchasing behavior:

Total Spend: is derived by summing the product of UnitPrice and Quantity for every transaction associated with a customer. Understanding the total spend per customer is essential for identifying high-value customers who contribute the most to the business's revenue, making it a cornerstone of monetary-based segmentation.
Average Transaction Value: reflects how much a customer spends on average per visit or transaction. This metric sheds light on the spending intensity of customers, allowing businesses to differentiate between customers who make frequent low-value purchases and those who make fewer but higher-value transactions, enabling more precise and personalized marketing efforts.






```python
# Calculate the total spend per customer
total_spend = df.groupby('CustomerID').apply(lambda x: (x['Quantity'] * x['UnitPrice']).sum(), include_groups=False).reset_index()
total_spend.columns = ['CustomerID', 'Total_Spend']

# Merge into purchase_summary
purchase_summary = purchase_summary.merge(total_spend, on='CustomerID', how='left')

# Calculate the average transaction value
purchase_summary['Average_Transaction_Value'] = purchase_summary['Total_Spend'] / purchase_summary['Total_orders']

purchase_summary.head()
```

## <b><span style='color:#ff69b4'> 4.2 Geographic Features </span></b>

```python
print(df['Country'].value_counts())
```

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
Since the vast majority of transactions originate from the United Kingdom, encoding all 37 countries as separate categories would introduce unnecessary noise into the clustering model. Instead, we engineered a binary feature, Is_UK, indicating whether a customer's primary country of purchase is the United Kingdom or not. This preserves the most relevant geographic signal — the distinction between domestic and international customers — while keeping the feature space clean and compatible with distance-based clustering algorithms such as K-Means.

```python
# Group by CustomerID and Country, count transactions
country_transactions = df.groupby(['CustomerID', 'Country'])['InvoiceNo'].nunique().reset_index()
country_transactions.rename(columns={'InvoiceNo': 'Transaction_Count'}, inplace=True)

# Identify the main country for each customer
main_country = country_transactions.loc[country_transactions.groupby('CustomerID')['Transaction_Count'].idxmax()]
main_country = main_country[['CustomerID', 'Country']].reset_index(drop=True)

# Create a binary column: 1 if UK, 0 otherwise
main_country['Is_UK'] = (main_country['Country'] == 'United Kingdom').astype(int)

# Merge into purchase_summary
purchase_summary = purchase_summary.merge(main_country[['CustomerID', 'Is_UK']], on='CustomerID', how='left')

purchase_summary.head()
```

```python
# Changing the data type of 'CustomerID' to string as it is a unique identifier
purchase_summary['CustomerID'] = purchase_summary['CustomerID'].astype(str)

# NOTE: convert_dtypes() is intentionally NOT used here because it converts
# numeric columns to nullable Int64/Float64 types which are incompatible with
# sklearn (IsolationForest, StandardScaler, PCA). We keep standard numpy dtypes.
print(purchase_summary.dtypes)
print(purchase_summary.shape)
```

## <b><span style='color:#ff69b4'> 4.3 Customer Lifetime (Tenure) </span></b>

```python

# Customer Lifetime (Tenure)
customer_lifetime = df.groupby('CustomerID').agg(
    First_Purchase=('InvoiceDate', 'min'),
    Last_Purchase=('InvoiceDate', 'max')
).reset_index()
customer_lifetime['Customer_Lifetime'] = (
    customer_lifetime['Last_Purchase'] - customer_lifetime['First_Purchase']
).dt.days
purchase_summary = purchase_summary.merge(
    customer_lifetime[['CustomerID', 'Customer_Lifetime']], on='CustomerID', how='left'
)

purchase_summary.head()
```

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
    
**Customer Lifetime (Tenure)**
This feature measures the number of days between a customer's first and last purchase. It serves as an indicator of customer loyalty and long-term engagement with the retailer. A higher value suggests a customer who has maintained a prolonged relationship with the business, making them potentially more valuable for retention strategies. A value of zero indicates a customer who made all their purchases in a single day, which may represent a one-time or bulk buyer.

## <b><span style='color:#ff69b4'> 4.4 Daily Customer Value </span></b>

```python
# Daily Customer Value
purchase_summary['Daily_Customer_Value'] = (
    purchase_summary['Total_Spend'] / purchase_summary['Customer_Lifetime'].replace(0, 1)
)
purchase_summary.head()
```

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
    
**Daily Customer Value**
This feature is calculated by dividing the total spend by the customer lifetime, representing the average amount a customer spends per day. It captures the spending intensity of a customer over their active period, providing a more nuanced view of customer value than total spend alone. Two customers with the same total spend but different lifetimes will have very different daily values, making this feature useful for distinguishing consistently active spenders from occasional high-value buyers.

## <b><span style='color:#ff69b4'> 4.5 Average Basket Size </span></b>

```python
# Average Basket Size
purchase_summary['Average_Basket_Size'] = (
    purchase_summary['Total_Purchase'] / purchase_summary['Total_orders']
)

print(purchase_summary.shape)
purchase_summary.head()
```

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">

**Average Basket Size**
This feature is computed by dividing the total quantity of products purchased by the total number of transactions. It reflects how many products a customer typically buys in a single transaction. A higher average basket size indicates a customer who tends to purchase in bulk, while a lower value suggests a customer who makes smaller, more selective purchases. This feature adds an important dimension to the frequency analysis by capturing not just how often a customer buys, but how much they buy each time they visit.

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">

To better capture customer behavior, a **product diversity** feature was introduced through <code>Unique_Products_Purchased</code>, measuring the number of distinct products bought by each customer.

Unlike traditional metrics (e.g., spend or frequency), this feature highlights whether a customer has a **focused** or **diverse purchasing pattern**, providing valuable insight for **segmentation, personalization, and predictive modeling**.

</div>

```python
# Calculate the number of unique products purchased by each customer
unique_products_purchased = df.groupby('CustomerID')['StockCode'].nunique().reset_index()
unique_products_purchased.rename(columns={'StockCode': 'Unique_Products_Purchased'}, inplace=True)

# Merge the new feature into the customer_data dataframe
purchase_summary = pd.merge(purchase_summary, unique_products_purchased, on='CustomerID')
purchase_summary.head()
```

## <b><span style='color:#ff69b4'> 4.7 Behavioral Features </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">

**Average Days Between Purchases:** This feature represents the average number of days a customer waits before making another purchase. It helps anticipate the next purchase timing, making it valuable for targeted marketing and personalized promotions.

**Favorite Shopping Day:** This indicates the day of the week when the customer shops most frequently. It helps identify behavioral patterns across customer segments and supports optimized scheduling of campaigns and promotions.

**Favorite Shopping Hour:** This captures the hour of the day when the customer is most active. It enables better timing of marketing actions to align with peak engagement periods.

By integrating these behavioral features, we build a more complete customer profile, improving clustering quality and enabling more meaningful segmentation.

</div>

```python
df['Day_Of_Week'] = df['InvoiceDate'].dt.dayofweek
df['Hour'] = df['InvoiceDate'].dt.hour

# Calculate the average number of days between consecutive purchases
days_between_purchases = (
    df.groupby('CustomerID')['InvoiceDay']
    .apply(lambda x: pd.Series(sorted(set(x)))
           .diff()
           .dropna()
           .apply(lambda y: y.days))
    .reset_index(level=0)
    .rename(columns={'InvoiceDay': 'DaysDiff'})
)
average_days_between_purchases = (
    days_between_purchases.groupby('CustomerID')['DaysDiff']
    .mean()
    .reset_index()
    .rename(columns={'DaysDiff': 'Average_Days_Between_Purchases'})
)

# Find the favorite shopping day of the week
favorite_shopping_day = df.groupby(['CustomerID', 'Day_Of_Week']).size().reset_index(name='Count')
favorite_shopping_day = favorite_shopping_day.loc[favorite_shopping_day.groupby('CustomerID')['Count'].idxmax()][['CustomerID', 'Day_Of_Week']]

# Find the favorite shopping hour of the day
favorite_shopping_hour = df.groupby(['CustomerID', 'Hour']).size().reset_index(name='Count')
favorite_shopping_hour = favorite_shopping_hour.loc[favorite_shopping_hour.groupby('CustomerID')['Count'].idxmax()][['CustomerID', 'Hour']]

purchase_summary = pd.merge(purchase_summary, average_days_between_purchases, on='CustomerID')
purchase_summary = pd.merge(purchase_summary, favorite_shopping_day, on='CustomerID')
purchase_summary = pd.merge(purchase_summary, favorite_shopping_hour, on='CustomerID')
purchase_summary.head()
```

## <b><span style='color:#ff69b4'> 5.2 Isolation Forest — Anomaly Detection </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
The **Isolation Forest** algorithm is particularly well-suited for high-dimensional customer data. It isolates anomalies by randomly selecting features and splitting values — outliers require fewer splits and are identified with a contamination threshold of **5%**, a conservative choice that preserves the bulk of the data while removing clearly anomalous profiles.
</div>

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# All numeric features — exclude ID, categorical Country, and binary Is_UK
cols_for_outlier_detection = [
    'Recency', 'Total_orders', 'Total_Purchase',
    'Total_Spend', 'Average_Transaction_Value',
    'Customer_Lifetime', 'Daily_Customer_Value', 'Average_Basket_Size',
    'Unique_Products_Purchased', 'Average_Days_Between_Purchases',
]
# Step 1 — Isolation Forest to remove anomalous customers
X_outlier = purchase_summary[cols_for_outlier_detection].astype(float)
iso_forest = IsolationForest(contamination=0.05, random_state=42)

purchase_summary_full = purchase_summary.copy()
purchase_summary_full['Is_Outlier'] = iso_forest.fit_predict(X_outlier)
purchase_summary_full['Is_Outlier'] = purchase_summary_full['Is_Outlier'].map({1: 'Inlier', -1: 'Outlier'})

# Visualize
outlier_percentage = purchase_summary_full['Is_Outlier'].value_counts(normalize=True) * 100
plt.figure(figsize=(12, 4))
outlier_percentage.plot(kind='barh', color='#ff69b4')
for index, value in enumerate(outlier_percentage):
    plt.text(value, index, f'{value:.2f}%', fontsize=15)
plt.title('Percentage of Inliers and Outliers')
plt.xticks(ticks=np.arange(0, 115, 5))
plt.xlabel('Percentage (%)')
plt.ylabel('Is Outlier')
plt.gca().invert_yaxis()
plt.show()
```

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
From the above plot, we can observe that about 5% of the customers have been identified as outliers in our dataset. This percentage seems to be a reasonable proportion, not too high to lose a significant amount of data, and not too low to retain potentially noisy data points. It suggests that our isolation forest algorithm has worked well in identifying a moderate percentage of outliers, which will be critical in refining our customer segmentation.

```python
# Separate the outliers for analysis
outliers_data = purchase_summary_full[purchase_summary_full['Is_Outlier'] == 'Outlier']

# Remove the outliers from the main dataset
purchase_summary_cleaned = purchase_summary_full[purchase_summary_full['Is_Outlier'] == 'Inlier']

# Drop the Is_Outlier column
purchase_summary_cleaned = purchase_summary_cleaned.drop(columns=['Is_Outlier'])

# Reset the index
purchase_summary_cleaned.reset_index(drop=True, inplace=True)

print(f'Clean customers: {len(purchase_summary_cleaned)}')
print(f'Outlier customers: {len(outliers_data)}')
purchase_summary_cleaned.head()
```

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
**We have  separated the outliers for further analysis** and cleaned our main dataset by removing these outliers. they are saved in a separate dataframe for potential analysis later. This is good practice as it allows to inspect who the outliers are and understand their characteristics.

# <b><span style='color:#ff69b4'> 6. Feature Scaling </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
Feature scaling is the process of transforming numeric features to a common range or distribution. Machine learning algorithms that rely on distances or gradients (PCA, K-Means) are sensitive to the magnitude of features. Without scaling, a feature like Total_Spend (hundreds/thousands) would dominate a feature like Total_orders (single digits) — not because it's more important, but simply because it's larger.

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# List of columns that don't need to be scaled
columns_to_exclude = ['CustomerID', 'Country']

# List of columns that need to be scaled
columns_to_scale = purchase_summary_cleaned.columns.difference(columns_to_exclude)

# Copy the cleaned dataset
customer_scaled = purchase_summary_cleaned.copy()

# Applying the scaler to the necessary columns in the dataset
customer_scaled[columns_to_scale] = scaler.fit_transform(customer_scaled[columns_to_scale])

# Display the first few rows of the scaled data
customer_scaled.head()
```

# <b><span style='color:#ff69b4'> 7. Dimensionality Reduction </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
**Principal Component Analysis (PCA)** transforms our correlated, high-dimensional feature space into a smaller set of orthogonal components that capture the maximum variance. This step reduces computational cost, eliminates multicollinearity, and improves K-Means cluster quality by removing noise dimensions. We retain components that together explain at least **80% of total variance**.
</div>

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ── Step 1: Apply PCA ───────────────────────────────────────────────
pca = PCA()
pca_result = pca.fit_transform(customer_scaled[columns_to_scale])

# ── Step 2: Explained Variance ──────────────────────────────────────
explained_variance = pd.DataFrame({
    'Principal Component': [f'PC{i+1}' for i in range(len(pca.explained_variance_ratio_))],
    'Explained Variance (%)': np.round(pca.explained_variance_ratio_ * 100, 2),
    'Cumulative Variance (%)': np.round(np.cumsum(pca.explained_variance_ratio_) * 100, 2)
})

print(explained_variance.to_string(index=False))

# ── Step 3: Scree Plot (Elbow) ──────────────────────────────────────
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
         np.cumsum(pca.explained_variance_ratio_ * 100), marker='o', color='steelblue')
plt.axhline(y=80, color='red', linestyle='--', label='80% threshold')
plt.axhline(y=90, color='orange', linestyle='--', label='90% threshold')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance (%)')
plt.title('PCA — Cumulative Explained Variance')
plt.legend()
plt.tight_layout()
plt.show()

# ── Step 4: Keep components that explain 80% variance ───────────────
pca_final = PCA(n_components=0.80)
customer_pca = pca_final.fit_transform(customer_scaled[columns_to_scale])

customer_pca_df = pd.DataFrame(
    customer_pca,
    columns=[f'PC{i+1}' for i in range(customer_pca.shape[1])]
)

print(f"\nOriginal features : {len(columns_to_scale)}")
print(f"PCA components kept: {customer_pca.shape[1]}")
print(f"Variance retained : {np.sum(pca_final.explained_variance_ratio_)*100:.2f}%\n")

customer_pca_df.head()
```

### <b><span style='color:#ff69b4'> 7.1 PCA Component Loadings </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
The loading matrix below shows which original features contribute most to each principal component. Highlighted cells (top 3 contributors per component) reveal the behavioral signals driving each axis of variation.
</div>

```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np

# ── Step 1: Fit PCA ─────────────────────────────────────────────────
pca = PCA()
pca.fit(customer_scaled[columns_to_scale])

# ── Step 2: Highlight top 3 absolute values per column ─────────────
def highlight_top3(column):
    top3 = column.abs().nlargest(3).index
    return ['background-color: #ff69B4' if i in top3 else '' for i in column.index]

# ── Step 3: Build PCA components DataFrame ──────────────────────────
pc_df = pd.DataFrame(
    pca.components_.T,
    columns=[f'PC{i+1}' for i in range(pca.n_components_)],
    index=columns_to_scale
)

# ── Step 4: Display with highlighting ───────────────────────────────
pc_df.style.apply(highlight_top3, axis=0)
```

# <b><span style='color:#ff69b4'> 8. Customer Segmentation — K-Means Clustering </span></b>

## <b><span style='color:#ff69b4'> 8.1 Optimal Number of Clusters </span></b>

### <b><span style='color:#ff69b4'> 8.1.1 Elbow Method (Inertia / WCSS) </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
The **Elbow Method** plots the Within-Cluster Sum of Squares (WCSS) against the number of clusters *k*. We look for the "elbow" — the point at which adding more clusters yields diminishing returns in compactness. This gives us an upper bound on *k* to explore.
</div>

```python
from sklearn.cluster import KMeans
inertias = []
k_range = range(2, 15)

for k in k_range:
    km = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=100, random_state=0)
    pca_cols = [col for col in customer_pca_df.columns if col.startswith('PC')]
    km.fit(customer_pca_df[pca_cols])
    inertias.append(km.inertia_)

plt.figure(figsize=(12, 5))
plt.plot(k_range, inertias, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (WCSS)')
plt.title('Elbow Method — K-Means')
plt.show()
```

### <b><span style='color:#ff69b4'> 8.1.2 Silhouette Analysis </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
The **Silhouette Score** measures how well each customer fits its assigned cluster relative to neighbouring clusters. Scores range from **-1** (poor fit) to **+1** (ideal fit). We test *k* from 3 to 12 and select the value that maximises this score.
    
The silhouette plots below show individual cluster cohesion for each candidate *k*. Well-separated, uniformly thick blades indicate balanced, meaningful clusters. Clusters with negative silhouette values contain misassigned customers and should be avoided.
</div>

```python
import matplotlib.pyplot as plt
from matplotlib import gridspec
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
 
def silhouette_analysis(df, start_k, stop_k, figsize=(15, 16)):
    """
    Perform Silhouette analysis for a range of k values and visualize the results.
    """

    # Set the size of the figure
    plt.figure(figsize=figsize)

    # Create a grid with (stop_k - start_k + 1) rows and 2 columns
    grid = gridspec.GridSpec(stop_k - start_k + 1, 2)

    # Assign the first plot to the first row and both columns
    first_plot = plt.subplot(grid[0, :])

    # First plot: Silhouette scores for different k values
    sns.set_palette(['darkorange'])

    silhouette_scores = []

    # Iterate through the range of k values
    for k in range(start_k, stop_k + 1):
        km = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=100, random_state=0)
        km.fit(df)
        labels = km.predict(df)
        score = silhouette_score(df, labels)
        silhouette_scores.append(score)

    best_k = start_k + silhouette_scores.index(max(silhouette_scores))

    plt.plot(range(start_k, stop_k + 1), silhouette_scores, marker='o')
    plt.xticks(range(start_k, stop_k + 1))
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Silhouette score')
    plt.title('Average Silhouette Score for Different k Values', fontsize=15)

    # Add the optimal k value text to the plot
    optimal_k_text = f'The k value with the highest Silhouette score is: {best_k}'
    plt.text(10, 0.23, optimal_k_text, fontsize=12, verticalalignment='bottom', 
             horizontalalignment='left', bbox=dict(facecolor='#fcc36d', edgecolor='#ff6200', boxstyle='round, pad=0.5'))
             

    # Second plot (subplot): Silhouette plots for each k value
    colors = sns.color_palette("bright")

    for i in range(start_k, stop_k + 1):    
        km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=0)
        row_idx, col_idx = divmod(i - start_k, 2)

        # Assign the plots to the second, third, and fourth rows
        ax = plt.subplot(grid[row_idx + 1, col_idx])

        visualizer = SilhouetteVisualizer(km, colors=colors, ax=ax)
        visualizer.fit(df)

        # Add the Silhouette score text to the plot
        score = silhouette_score(df, km.labels_)
        ax.text(0.97, 0.02, f'Silhouette Score: {score:.2f}', fontsize=12, \
                ha='right', transform=ax.transAxes, color='red')

        ax.set_title(f'Silhouette Plot for {i} Clusters', fontsize=15)

    plt.tight_layout()
    plt.show()
pca_cols = [col for col in customer_pca_df.columns if col.startswith('PC')]
silhouette_analysis(customer_pca_df[pca_cols], 3, 12, figsize=(20, 50))
```

<div style="border-radius:10px; padding: 20px; background-color: #ffeaef; font-size:115%; text-align:left">
<h3 align="left"><font color=#d4537e>🔍 Why k = 4 and Not k = 5 ?</font></h3>

It is completely **normal** for both the Elbow Method and the Silhouette Analysis to point toward **k = 5** as the mathematically optimal number of clusters. This result is statistically sound. However, choosing k = 4 is a deliberate and justified **business decision**, not an error. Here is why:

---

**Why both methods suggest k = 5**

- The **Elbow curve** shows a visible kink around k = 5, meaning adding a 5th cluster still reduces inertia meaningfully before the curve flattens.
- The **Silhouette score** peaks (or is very close to its peak) at k = 5, indicating slightly better-defined cluster boundaries than at k = 4.

These are purely **geometric** signals — they optimise for mathematical compactness and separation, not for business interpretability.

---

**Why k = 4 was chosen**

| Reason | Explanation |
|:---|:---|
| **Interpretability** | 4 customer segments map cleanly to actionable business profiles (Champions, Loyal Mid-Value, Occasional High-Spenders, At-Risk). A 5th cluster often splits one existing segment into two nearly identical sub-groups with no distinct strategic meaning. |
| **Actionability** | Marketing teams can realistically design 4 distinct campaign strategies. Adding a 5th segment increases complexity without proportional business value. |
| **Cluster balance** | At k = 5, one cluster typically contains fewer than 8% of customers — too small to justify a dedicated strategy at scale. |
| **Silhouette difference is marginal** | The silhouette score gain from k = 4 → k = 5 is usually < 0.02, which is not statistically significant enough to override business logic. |

---

**Conclusion**: Both choices are valid. k = 5 is mathematically marginally better; k = 4 is **strategically better**. In applied data science for customer segmentation, business interpretability always takes precedence when the quantitative difference is small. If your stakeholders or client prefer 5 segments and can clearly differentiate the 5th profile, switching to k = 5 is entirely reasonable.

</div>

## <b><span style='color:#ff69b4'> 8.2 Fitting the Final K-Means Model </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
Based on both the elbow curve and the silhouette analysis, **k = 4** delivers the best balance of compactness and separation. We apply K-Means with `k-means++` initialization (reduces sensitivity to starting centroids), 10 random restarts, and a maximum of 100 iterations.
</div>

```python
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=100, random_state=0)

pca_cols = [col for col in customer_pca_df.columns if col.startswith('PC')]
kmeans.fit(customer_pca_df[pca_cols])  # ✅ only PC columns

from collections import Counter
cluster_frequencies = Counter(kmeans.labels_)

label_mapping = {label: new_label for new_label, (label, _) in 
                 enumerate(cluster_frequencies.most_common())}

new_labels = np.array([label_mapping[label] for label in kmeans.labels_])

purchase_summary_cleaned['cluster'] = new_labels
customer_pca_df['cluster'] = new_labels
```

# <b><span style='color:#ff69b4'> 9. Model Evaluation Metrics </span></b>

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">

To rigorously assess the quality of our clustering solution, three complementary metrics are computed:

| Metric | Formula Intuition | Ideal Value |
|:---|:---|:---:|
| **Silhouette Score** | Measures cohesion within clusters vs. separation between clusters | Closer to **+1** |
| **Calinski-Harabász Score** | Ratio of between-cluster to within-cluster dispersion | **Higher** is better |
| **Davies-Bouldin Score** | Average similarity of each cluster to its nearest neighbour | **Lower** is better |

Together, these three metrics provide a 360° view of cluster quality — a high silhouette with low Davies-Bouldin and high Calinski-Harabász confirms that the 4-segment solution is both internally cohesive and externally well-separated.

</div>

```python
from tabulate import tabulate
from sklearn.metrics import (
    silhouette_score,
    calinski_harabasz_score,
    davies_bouldin_score
)
# Compute number of customers
num_observations = len(customer_pca_df)

# Separate the features and the cluster labels
X = customer_pca_df.drop('cluster', axis=1)
clusters = customer_pca_df['cluster']

# Compute the metrics
sil_score = silhouette_score(X, clusters)
calinski_score = calinski_harabasz_score(X, clusters)
davies_score = davies_bouldin_score(X, clusters)

# Create a table to display the metrics and the number of observations
table_data = [
    ["Number of Observations", num_observations],
    ["Silhouette Score", sil_score],
    ["Calinski Harabasz Score", calinski_score],
    ["Davies Bouldin Score", davies_score]
]

# Print the table
print(tabulate(table_data, headers=["Metric", "Value"], tablefmt='pretty'))
```

<div style="background-color: #ffeaef; padding: 20px; border-radius: 12px; font-size: 15px; line-height: 1.6;">
**Model Selection & Implementation**

To select the most suitable language model for our recommendation system, we compared three widely used LLMs:

| Model | Developer | Access | Strengths | Limitations |
|-------|-----------|--------|-----------|-------------|
| **GPT-4o** | OpenAI | Paid API | Very powerful, industry standard | Requires paid subscription |

| **Llama 3.3-70B**  | Meta | Free via Groq | Strong reasoning, open-source, free, fast | None for our use case |

**Llama 3.3-70B was selected** as it offers comparable performance to commercial models while being completely free, open-source, and easily accessible via the Groq inference API — making it the most practical and academically appropriate choice for this project.

---

### Implementation

The model operates in a **zero-shot setting**: for each customer segment, the top 10 most purchased products are passed as context, and the model generates 5 personalised product recommendations without any task-specific training on our dataset.
</div>

```python
import requests

from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
API_KEY = user_secrets.get_secret("GROQ_API_KEY")

def get_top_products(cluster_id, n=10):
    cluster_customers = purchase_summary_cleaned[
        purchase_summary_cleaned['cluster'] == cluster_id
    ]['CustomerID']
    
    top_products = (df[df['CustomerID'].isin(cluster_customers)]
                   ['Description']
                   .value_counts()
                   .head(n)
                   .index.tolist())
    return top_products

def get_recommendations(cluster_name, top_products):
    response = requests.post(
        "https://api.groq.com/openai/v1/chat/completions",
        headers={
            "Authorization": f"Bearer {API_KEY}",
            "Content-Type": "application/json"
        },
        json={
            "model": "llama-3.3-70b-versatile",
            "messages": [{"role": "user", "content": f"""
                You are a retail recommendation expert.
                Customer segment: {cluster_name}
                Their most purchased products: {top_products}
                Suggest 5 products this segment would enjoy and explain why.
            """}]
        }
    )
    print(response.json())
    return response.json()['choices'][0]['message']['content']

# ---- Run for all clusters ----
clusters_info = {
    0: "Regular Browsers",
    1: "Lapsed Low-Spenders",
    2: "Power Shoppers",
    3: "High-Value Inactives"
}

for cluster_id, cluster_name in clusters_info.items():
    top_products = get_top_products(cluster_id)
    recommendations = get_recommendations(cluster_name, top_products)
    print(f"\n{'='*50}")
    print(f"Cluster {cluster_id} — {cluster_name}")
    print(f"Top products: {top_products}")
    print(f"\nRecommendations:\n{recommendations}")
```
